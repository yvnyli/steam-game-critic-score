# Predict probability of a game receiving a Metacritic score
### Purpose
To estimate credibility of predicted score. Games with a high predicted probability are more similar to games with scores, and thus their predicted scores are more credible.
### Method
Use ridge regression because we want regularization that is stable (ridge is more stable than lasso).
### Notes
1. Use all data (n=83560)
2. Check that the probability is well-calibrated (among the games with 80-90% predicted probability, indeed 80-90% of them actually have a score)
3. Cross-validation to avoid overfitting
4. Check stability over random seeds

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
def select_C(X, y, Cs):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
    scores = {}

    for C in Cs:
        fold_scores = []
        for tr, va in skf.split(X, y):
            model = LogisticRegression(
                C=C,
                l1_ratio=0,
                max_iter=500,
                random_state=0,
            )
            model.fit(X.iloc[tr], y.iloc[tr])
            p = model.predict_proba(X.iloc[va])[:, 1]
            fold_scores.append(log_loss(y.iloc[va], p))
        scores[C] = sum(fold_scores) / len(fold_scores)

    best_C = min(scores, key=scores.get)
    return best_C, scores

best_C, scores = select_C(df_X_z, df_y_z["has_score"],  Cs=(0.01,0.03,0.1,0.3,1,3,10))
print(best_C)
print(scores)
```

```python
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import brier_score_loss, log_loss, roc_auc_score
from scipy.stats import spearmanr

def expected_calibration_error(y_true, p_pred, n_bins=15, strategy="quantile"):
    """
    ECE with either quantile bins (roughly equal counts) or uniform bins.
    """
    y_true = np.asarray(y_true).astype(int)
    p_pred = np.asarray(p_pred).astype(float)

    if strategy == "quantile":
        # Bin edges by quantiles of predictions
        quantiles = np.linspace(0, 1, n_bins + 1)
        bin_edges = np.quantile(p_pred, quantiles)
        # Ensure monotonic edges (can have duplicates if many identical probs)
        bin_edges[0], bin_edges[-1] = 0.0, 1.0
        bin_edges = np.unique(bin_edges)
    elif strategy == "uniform":
        bin_edges = np.linspace(0, 1, n_bins + 1)
    else:
        raise ValueError("strategy must be 'quantile' or 'uniform'")

    # Assign bins
    # np.digitize returns 1..len(bin_edges)-1
    bin_ids = np.digitize(p_pred, bin_edges[1:-1], right=True)

    ece = 0.0
    n = len(y_true)

    for b in np.unique(bin_ids):
        idx = np.where(bin_ids == b)[0]
        if len(idx) == 0:
            continue
        acc = y_true[idx].mean()
        conf = p_pred[idx].mean()
        ece += (len(idx) / n) * abs(acc - conf)

    return float(ece)
    
def train_similarity_model_with_calibration(X_train, y_train, base_C=1.0, n_splits=5, calibrate=True, method="isotonic", random_state=0):
    """
    Base model = logistic regression (interpretable)
    Then optionally calibrate using CalibratedClassifierCV.
    """
    base = LogisticRegression(
        C=base_C,
        l1_ratio=0,
        max_iter=500,
        random_state=random_state,
    )

    if calibrate:
        skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    
        # CalibratedClassifierCV uses an internal CV on the training set
        # to fit a calibration map. This avoids leaking the external test set.
        cal = CalibratedClassifierCV(
            estimator=base,
            method=method,     # "sigmoid" (Platt) or "isotonic"
            cv=skf
        )
        cal.fit(X_train, y_train)
        return cal
    else:
        base.fit(X_train, y_train)
        return base

def evaluate_calibration(y_true, p_pred, n_bins=15):
    y_true = np.asarray(y_true).astype(int)
    p_pred = np.asarray(p_pred).astype(float)

    metrics = {}
    metrics["prevalence"] = float(y_true.mean())
    metrics["mean_pred"] = float(p_pred.mean())
    metrics["brier"] = float(brier_score_loss(y_true, p_pred))
    metrics["log_loss"] = float(log_loss(y_true, np.clip(p_pred, 1e-15, 1-1e-15)))
    metrics["auc"] = float(roc_auc_score(y_true, p_pred))
    metrics["ece_quantile"] = expected_calibration_error(y_true, p_pred, n_bins=n_bins, strategy="quantile")
    metrics["ece_uniform"] = expected_calibration_error(y_true, p_pred, n_bins=n_bins, strategy="uniform")
    return metrics

def stability_via_repeats(X, y, base_C=1.0, n_repeats=5, test_size=0.2, random_seed=0,
                          calibrate=True, calib_method="isotonic"):
    """
    Re-train with different random splits; collect probability predictions on a fixed held-out set.
    Measure variability + rank stability.
    """
    # Fixed held-out set for stability evaluation
    X_tr, X_te, y_tr, y_te = train_test_split(
        X, y, test_size=test_size, stratify=y, random_state=random_seed
    )

    P = []  # list of (n_test,) arrays
    cal_metrics = []

    for r in range(n_repeats):
        rs = random_seed + r + 1
        # Re-split training set a bit to emulate resampling variability
        X_tr_r, X_val_r, y_tr_r, y_val_r = train_test_split(
            X_tr, y_tr, test_size=0.25, stratify=y_tr, random_state=rs
        )

        model = train_similarity_model_with_calibration(
            X_tr_r, y_tr_r, base_C=base_C, calibrate=calibrate, method=calib_method, random_state=rs
        )

        p_te = model.predict_proba(X_te)[:, 1]
        P.append(p_te)
        cal_metrics.append(evaluate_calibration(y_te, p_te))

    P = np.vstack(P)  # shape (n_repeats, n_test)
    mu = P.mean(axis=0)
    sd = P.std(axis=0, ddof=1)

    # Stability summaries
    out = {
        "mean_sd_all": float(sd.mean()),
        "median_sd_all": float(np.median(sd)),
        "mean_sd_mu_gt_0p7": float(sd[mu > 0.7].mean()) if np.any(mu > 0.7) else np.nan,
        "frac_mu_gt_0p7": float((mu > 0.7).mean()),
    }

    # Rank stability (Spearman correlation) across repeats
    corrs = []
    for a in range(n_repeats):
        for b in range(a + 1, n_repeats):
            corrs.append(spearmanr(P[a], P[b]).correlation)
    out["spearman_mean"] = float(np.nanmean(corrs))
    out["spearman_min"] = float(np.nanmin(corrs))

    # Also return calibration metrics across repeats (mean)
    keys = cal_metrics[0].keys()
    out["calibration_mean"] = {k: float(np.mean([cm[k] for cm in cal_metrics])) for k in keys}
    out["calibration_std"] = {k: float(np.std([cm[k] for cm in cal_metrics], ddof=1)) for k in keys}

    return out

# Example usage:
sim_report = stability_via_repeats(df_X_z, df_y_z["has_score"], base_C=best_C, n_repeats=5, calibrate=True, calib_method="isotonic")
import pprint
pprint.pp(sim_report)
```

```python
from sklearn.calibration import calibration_curve

def plot_calibration_curve(y_true, p_pred, n_bins=15, strategy="uniform"):
    """
    strategy:
      - "quantile": equal number of samples per bin (recommended for imbalance)
      - "uniform": equal-width bins in [0,1]
    """
    frac_pos, mean_pred = calibration_curve(y_true, p_pred, n_bins=n_bins,
        strategy=strategy)

    plt.figure(figsize=(4, 4))
    plt.plot(mean_pred, frac_pos, "o-", label="Model")
    plt.plot([0, 1], [0, 1], "--", color="gray", label="Perfect calibration")

    plt.xlabel("Mean predicted probability")
    plt.ylabel("Fraction with field present")
    plt.title(f"Calibration curve ({strategy} bins, n_bins={n_bins})")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_calibration_curve(y_true, p_pred, n_bins=15, strategy="uniform")
```

## Is this prediction a reasonable one? Yes
1. The performance is good: mean_pred matches true prevalence, Brier score loss and log loss are low, ROC AUC is high, expected calibration errors are low.
2. It is stable across random seeds: small sd and std in all metrics, correlation of pred across random seeds is high.
3. It's well calibrated: see calibration curve.

```python
def plot_probability_cdf_with_refs(y_true, p_pred, thresholds=(0.1, 0.3, 0.4, 0.5, 0.7)):
    p = np.asarray(p_pred)
    p_sorted = np.sort(p)
    y = np.arange(1, len(p_sorted) + 1) / len(p_sorted)

    plt.figure(figsize=(6, 4))
    plt.plot(p_sorted, y, label="CDF")

    for t in thresholds:
        frac = (p > t).mean()
        y_true_frac = y_true[p>t].sum() / y_true.sum()
        plt.axvline(t, linestyle="--", alpha=0.5)
        plt.text(t, 0.02, f"{frac:.1%} > {t}, {y_true_frac:.1%} of y=1", rotation=90,
                 va="bottom", ha="right", fontsize=9)

    plt.xlabel("Predicted probability p")
    plt.ylabel("Fraction of samples  p")
    plt.title("CDF of coverage probabilities\n(vertical lines = trust thresholds)")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
plot_probability_cdf_with_refs(y_true, p_pred)
```

```python
def plot_p_pred_histograms(p_pred, y_true, bins=50, thresholds=(0, 0.1, 0.4, 0.7)):
    p_pred = np.asarray(p_pred)
    y_true = np.asarray(y_true)

    p_pos = p_pred[y_true == 1]
    p_neg = p_pred[y_true == 0]

    plt.figure(figsize=(6, 4))
    
    plt.hist(p_neg, bins=bins, density=True, alpha=0.6,
        label="y = 0 (no Metacritic score)")

    plt.hist(p_pos, bins=bins, density=True, alpha=0.6,
        label="y = 1 (has Metacritic score)")

    for t in thresholds:
        y_true_frac = y_true[p_pred>t].mean()
        plt.axvline(t, linestyle="--", alpha=0.5)
        plt.text(t, 0.02, f"      {y_true_frac:.1%} are y=1 with > {t}", rotation=90,
                 va="bottom", ha="right", fontsize=9)
        
    plt.xlabel("Predicted probability p")
    plt.ylabel("Density")
    plt.title("Distribution of predicted probabilities")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
plot_p_pred_histograms(p_pred, y_true)
```
